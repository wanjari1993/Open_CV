```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
datagen=ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
Image=load_img(r"D:\Naresh IT\Class Notes\Completed\Deep Learning (DL)\Happy-Sad Face Images\DOG.jpeg")
```

```python
Image
```

```python
x=img_to_array(Image)
x=x.reshape((1,) + x.shape)

i=0
for batch in datagen.flow(x, batch_size=1,
                         save_to_dir=r"D:\Naresh IT\Class Notes\Completed\Deep Learning (DL)\Happy-Sad Face Images\Sad-Validation", save_prefix='cat', save_format='jpeg'):
    i +=1
    if i>30:
        break 
```
